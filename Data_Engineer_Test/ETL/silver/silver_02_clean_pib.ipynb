```scala
#Leitura do arquivo parquet
df = spark.read.parquet("bronze/ingest_ibge_pib")
df = df.orderBy("createdAt", ascending=False).limit(1) #Pegar só o mais recente

df.show(truncate=36) #Coloquei 36 só para mostrar o id sem corte
```

```scala
#Faz a leitura do json e captura dos dados necessários
raw_id = df.select("id").collect()[0][0]

schema = df.select(schema_of_json(df.select("json").first()[0]).alias("schema")).first()["schema"] #faço um schema "automatico" com o schema_of_json 


df_final = (
    df
    .withColumn("json_parsed", from_json("json", schema)) #cria a coluna parseando o json
    .withColumn("item", explode("json_parsed")) #explode o campo item
    .withColumn("resultado", explode("item.resultados")) #explode o campo resultado
    .withColumn("serie_item", explode("resultado.series")) #explode o campo series
    .select(
        col("serie_item.localidade.nome").alias("localidade"), #pega a coluna final localidade
        col("serie_item.serie")["2021"].cast("double").alias("pib_2021") #pega a coluna final pib
    )
)

df_final.show()
```

```scala
#Aplica filtros
df_final = df_final.withColumn("cidade", expr("trim(substring_index(localidade, ' - ', 1))")) #pega tudo a esquerda de ' - '
df_final = df_final.withColumn("estado", expr("rtrim(right(localidade, 2))")) # pega os estados
df_final.show(truncate=False)
```

```scala
#Cria as colunas necessárias
createdAt = datetime.now() #pega o horário de agora
deletedAt = None #define padrao null

df_final = df_final.withColumn("id", expr("uuid()")) #gera uuids
df_final = df_final.withColumn("raw_id", lit(raw_id)) #insere o raw_id do json para rastreabilidade
df_final = df_final.withColumn("createdAt", lit(createdAt))
df_final = df_final.withColumn("deletedAt", lit(deletedAt).cast(TimestampType()))

df = df.withColumn("createdAt", from_utc_timestamp("createdAt", "America/Sao_Paulo")) #Altera o timestamp BR

df_final = df_final.select(['id', 'cidade', 'estado', 'pib_2021', 'createdAt', 'deletedAt', 'raw_id']) #seleciona as colunas na ordem correta
```

```scala
#Pegando dados de estado e UF que já tinha salvo para evitar recriar
df1 = spark.read.parquet("silver/clean_populacao_estado") #leitura dos estados
df2 = spark.read.parquet("silver/clean_populacao_cidade") #leitura das cidades
df2 = df2.withColumnRenamed("id", "id2") #renomeando para nao bugar o join


df_joined = df2.join(df1, df2["id_estado"] == df1["id"], "left") #join para pegar os ids de cidade/estado
df_joined = df_joined.select(['id', 'id2', 'cidade', 'estado'])

df_joined = df_joined.withColumnRenamed("id", "id_estado") #renomeando
df_joined = df_joined.withColumnRenamed("id2", "id_cidade")
df_joined = df_joined.withColumnRenamed("cidade", "cidade2")
df_joined = df_joined.withColumnRenamed("estado", "estado2")

#join para o pib se relacionar com a cidade cujo id já foi criado
df_joined = df_joined.join(df_final, (df_joined["cidade2"] == df_final["cidade"]) & (df_joined["estado2"] == df_final["estado"]), "left")
```

```scala
#Cria tabela fato pib
df_final = df_joined.select(['id', 'pib_2021', 'createdAt', 'deletedAt', 'id_cidade'])

df_final.write.mode("overwrite").parquet("silver/clean_pib_pib") #Salvando em parquet 
```
