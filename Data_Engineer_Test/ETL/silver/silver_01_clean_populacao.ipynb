```scala
df = spark.read.parquet("bronze/ingest_ibge_populacao")
df = df.orderBy("createdAt", ascending=False).limit(1) #Pegar só o mais recente

df.show(truncate=36) #Coloquei 36 só para mostrar o id sem corte
```

```scala
raw_id = df.select("id").collect()[0][0]

schema = df.select(schema_of_json(df.select("json").first()[0]).alias("schema")).first()["schema"] #faço um schema "automatico" com o schema_of_json 

df_final = (
    df
    .withColumn("json_parsed", from_json("json", schema)) #cria a coluna parseando o json
    .withColumn("item", explode("json_parsed")) #explode o campo item
    .withColumn("resultado", explode("item.resultados")) #explode o campo resultado
    .withColumn("serie_item", explode("resultado.series")) #explode o campo series
    .select(
        col("serie_item.localidade.nome").alias("localidade"), #pega a coluna final localidade
        col("serie_item.serie")["2024"].cast("int").alias("qtd_populacao") #pega a coluna final quantidade população
    )
)
```

```scala
df_final = df_final.withColumn("cidade", expr("trim(substring_index(localidade, ' - ', 1))")) #pega tudo a esquerda de ' - '
df_final = df_final.withColumn("estado", expr("rtrim(right(localidade, 2))")) # pega os estados
df_final.show(truncate=False)
```

```scala
createdAt = datetime.now() #pega o horário de agora
deletedAt = None #define padrao null


df_final = df_final.withColumn("id", expr("uuid()")) #gera uuids
df_final = df_final.withColumn("raw_id", lit(raw_id)) #insere o raw_id do json para rastreabilidade
df_final = df_final.withColumn("createdAt", lit(createdAt))
df_final = df_final.withColumn("deletedAt", lit(deletedAt).cast(TimestampType()))

df_final = df_final.withColumn("createdAt", from_utc_timestamp("createdAt", "America/Sao_Paulo")) #Altera o timestamp BR


df_final = df_final.select(['id', 'cidade', 'estado', 'qtd_populacao', 'createdAt', 'deletedAt', 'raw_id']) #seleciona as colunas na ordem correta
```

```scala
df1 = df_final.select("estado").distinct() #Tiro os estados duplicados
df1 = df1.withColumn("id", expr("uuid()")) #Adiciono as mesmas informações de antes, porém na tabela com estados distintos
df1 = df1.withColumn("raw_id", lit(raw_id))
df1 = df1.withColumn("createdAt", lit(createdAt))
df1 = df1.withColumn("deletedAt", lit(deletedAt).cast(TimestampType()))
df1 = df1.withColumn("createdAt", from_utc_timestamp("createdAt", "America/Sao_Paulo")) #Altera o timestamp BR
df1 = df1.select(['id', 'estado', 'createdAt', 'deletedAt', 'raw_id'])
df1 = df1.withColumnRenamed('id', 'id1') #Faço a renomeação das colunas para não dar erro nos joins
df1 = df1.withColumnRenamed('estado', 'estado1')
df1 = df1.withColumnRenamed('createdAt', 'createdAt1')
df1 = df1.withColumnRenamed('deletedAt', 'deletedAt1')

df2 = df_final.join(df1, df_final["estado"] == df1["estado1"], "left") #join para trazer os ids dos estados para a tabela de cidades
df2 = df2.withColumnRenamed('id1', 'id_estado')
df2 = df2.withColumn("id", expr("uuid()"))
df2 = df2.select(['id', 'cidade', 'createdAt', 'deletedAt', 'id_estado', 'qtd_populacao']) #mantenho qtd_populacao para inserir na tabela 3

df3 = df2.withColumnRenamed("id", "id_cidade")
df3 = df3.withColumn("id", expr("uuid()"))

df1 = df1.withColumnRenamed('id1', 'id')
df1 = df1.withColumnRenamed('estado1', 'estado')
df1 = df1.withColumnRenamed('createdAt1', 'createdAt')
df1 = df1.withColumnRenamed('deletedAt1', 'deletedAt')

#Seleção dos dados finais em cada tabela
df1 = df1.select(['id', 'estado', 'createdAt', 'deletedAt', 'raw_id'])
df2 = df2.select(['id', 'cidade', 'createdAt', 'deletedAt', 'id_estado'])
df3 = df3.select(['id', 'qtd_populacao', 'createdAt', 'deletedAt', 'id_cidade'])

df1.write.mode("overwrite").parquet("silver/clean_populacao_estado") #Salvando em parquet 
df2.write.mode("overwrite").parquet("silver/clean_populacao_cidade")
df3.write.mode("overwrite").parquet("silver/clean_populacao_habitantes") 
# #Aqui a ideia é salvar os dados em uma tabela mantendo a rastreabilidade para saber de onde vieram
```
