```scala
df_estado = spark.read.parquet("silver/clean_populacao_estado").orderBy(col("createdAt").desc())
df_cidade = spark.read.parquet("silver/clean_populacao_cidade").orderBy(col("createdAt").desc())
df_habitantes = spark.read.parquet("silver/clean_populacao_habitantes").orderBy(col("createdAt").desc())
df_pib = spark.read.parquet("silver/clean_pib_pib").orderBy(col("createdAt").desc())
```

```scala
df_estado = df_estado.filter("deletedAt IS NULL")
df_cidade = df_cidade.filter("deletedAt IS NULL")
df_habitantes = df_habitantes.filter("deletedAt IS NULL")
df_pib = df_pib.filter("deletedAt IS NULL")
```

```scala
df_estado_ativo = (df_estado.filter(col("deletedAt").isNull())) #tira dados inválidos (null se houver)
df_estado_ativo = df_estado_ativo.withColumnRenamed("id", "id_estado")
df_estado_ativo = df_estado_ativo.withColumnRenamed("estado", "uf")
df_estado_ativo = df_estado_ativo.withColumnRenamed("createdAt", "estado_createdAt")
df_estado_ativo = df_estado_ativo.withColumnRenamed("raw_id", "estado_raw_id")

df_cidade_ativa = (df_cidade.filter(col("deletedAt").isNull()))
df_cidade_ativa = df_cidade_ativa.withColumnRenamed("id", "id_cidade")
df_cidade_ativa = df_cidade_ativa.withColumnRenamed("cidade", "nome_cidade")
df_cidade_ativa = df_cidade_ativa.withColumnRenamed("createdAt", "cidade_createdAt")

df_habitantes_ativo = (df_habitantes.filter(col("deletedAt").isNull()))
df_habitantes_ativo = df_habitantes_ativo.withColumnRenamed("id", "id_habitantes")
df_habitantes_ativo = df_habitantes_ativo.withColumnRenamed("qtd_populacao", "qtd_populacao")
df_habitantes_ativo = df_habitantes_ativo.withColumnRenamed("createdAt", "habitantes_createdAt")
df_habitantes_ativo = df_habitantes_ativo.withColumn("qtd_populacao", col("qtd_populacao").cast("double"))

df_pib_ativo = (df_pib.filter(col("deletedAt").isNull()))
df_pib_ativo = df_pib_ativo.withColumnRenamed("id", "id_pib")
df_pib_ativo = df_pib_ativo.withColumnRenamed("pib_2021", "pib")
df_pib_ativo = df_pib_ativo.withColumnRenamed("createdAt", "pib_createdAt")
df_pib_ativo = df_pib_ativo.withColumn("pib", col("pib").cast("double"))
```

```scala
df_join1 = df_cidade_ativa.alias("cid").join(df_estado_ativo.alias("est"), on="id_estado", how="inner")

df_join2 = df_join1.join(df_habitantes_ativo.alias("hab"), on="id_cidade", how="inner")

df_join3 = df_join2.join(df_pib_ativo.alias("pib"), on="id_cidade", how="inner")

#Selecionando as colunas que vou usar evitando as ambiguidades 
df_base = df_join3.select(
    col("id_cidade"),
    col("nome_cidade"),
    col("uf"),
    lit(2021).alias("ano"),
    col("qtd_populacao").alias("POPULACAO_QTD"),
    col("pib"),
    col("id_estado"),
    col("id_habitantes"),
    col("id_pib"),
    col("estado_createdAt"),
    col("cidade_createdAt"),
    col("habitantes_createdAt"),
    col("pib_createdAt"),
    col("estado_raw_id")
)
```

```scala
df_base = df_base.filter("POPULACAO_QTD IS NOT NULL AND POPULACAO_QTD > 0 AND pib IS NOT NULL AND uf IS NOT NULL")
```

```scala
w_uf = Window.partitionBy("uf", "ano")
w_brasil = Window.partitionBy("ano")

w_rank_uf = Window.partitionBy("uf", "ano").orderBy(
    col("pib").desc(), col("nome_cidade").asc(), col("id_cidade").asc())

w_rank_brasil = Window.partitionBy("ano").orderBy(
    col("pib").desc(), col("uf").asc(), col("nome_cidade").asc(), col("id_cidade").asc())
```

```scala
df_pib_indicadores = df_base.withColumn("pib_per_capita", expr("pib * 1000 / POPULACAO_QTD"))
df_pib_indicadores = df_pib_indicadores.withColumn("pib_total_uf_ano", sum("pib").over(w_uf))
df_pib_indicadores = df_pib_indicadores.withColumn("pib_total_brasil_ano", sum("pib").over(w_brasil))
df_pib_indicadores = df_pib_indicadores.withColumn("share_uf", expr("pib / pib_total_uf_ano"))
df_pib_indicadores = df_pib_indicadores.withColumn("share_brasil", expr("pib / pib_total_brasil_ano"))
df_pib_indicadores = df_pib_indicadores.withColumn("ranking_uf", dense_rank().over(w_rank_uf))
df_pib_indicadores = df_pib_indicadores.withColumn("ranking_brasil", dense_rank().over(w_rank_brasil))

df_pib_indicadores = df_pib_indicadores.select([
    'nome_cidade', 'uf', 'ano', 'pib',
    'POPULACAO_QTD', 'pib_per_capita', 'share_uf', 'share_brasil',
    'ranking_uf', 'ranking_brasil'
    ]) 
    
df_uf = (df_pib_indicadores.groupBy("uf", "ano").agg(
          sum("pib").alias("pib_total"),
          sum("POPULACAO_QTD").alias("populacao_total"),
          avg("pib_per_capita").alias("pib_per_capita_medio")).orderBy("uf", "ano"))
          
df_uf = df_uf.select([
    'uf', 'ano', 'populacao_total', 
    'pib_total', 'pib_per_capita_medio']) 
    
    
df_brasil = (df_pib_indicadores.groupBy("ano").agg(
        sum("pib").alias("pib_total_brasil"),
        sum("POPULACAO_QTD").alias("populacao_total_brasil"),
        avg("pib_per_capita").alias("pib_per_capita_medio_brasil"),
        max("share_uf").alias("share_uf_max"),
        max("share_brasil").alias("share_brasil_max"),
        max("ranking_uf").alias("ranking_uf_top"),
        max("ranking_brasil").alias("ranking_brasil_top")).orderBy("ano"))

#Não consegui calcular YoY pois pelo q pesquisei é necessário pip do ano anterior e fiz a ingestão apenas de 2021.
```

```scala
#1.	Município-ano (pib, população, pib per capita, YoY, share na UF e no Brasil, rankings);
df_pib_indicadores.show(truncate=False)

#2.	UF-ano (agregado por UF, idem métricas);
df_uf.show(truncate=False)

#3.	Brasil-ano (agregado nacional).
df_brasil.show(truncate=False)
```

```scala
df_pib_indicadores.write.mode("overwrite").parquet("gold/indicadores_municipio") #Salvando em parquet 
df_uf.write.mode("overwrite").parquet("gold/indicadores_uf")
df_brasil.write.mode("overwrite").parquet("gold/indicadores_brasil") 
```
