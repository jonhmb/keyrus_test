```scala
schema_ingest_ibge_pib = StructType([
    StructField("id", StringType()),
    StructField("json", StringType()),
    StructField("createdAt", TimestampType())
])
```

```scala
url = "https://servicodados.ibge.gov.br/api/v3/agregados/5938/periodos/-6/variaveis?localidades=BR" #de acordo com a documentacao """https://servicodados.ibge.gov.br/api/docs/agregados?versao=3""" é necessário montar a URL desse jeito para obter a lista de variáveis
response = requests.get(url).json()

print(response) #json formado

#neste caso a variável que preciso é a 37 'Produto Interno Bruto a preços correntes'
```

```scala
url = "https://servicodados.ibge.gov.br/api/v3/agregados/5938/periodos/2021/variaveis/37?localidades=N6" #montando a URL de request com a variável do PIB (n6 = municipios)

response = requests.get(url).json()

print(response) #json formado
```

```scala
id = uuid.uuid4() #retorna id tipo uuidv4
response = json.dumps(response) #deixa o json da api em formato válido
hora = datetime.now() #pega o horário de agora

dados = [(str(id), response, hora)]

df = spark.createDataFrame(dados, schema_ingest_ibge_pib) #cria o dataframe com os dados necessários e o schema indicado anteriormente

df = df.withColumn("createdAt", from_utc_timestamp("createdAt", "America/Sao_Paulo")) #Altera o timestamp BR

df.write.mode("overwrite").parquet("bronze/ingest_ibge_pib") #Salvando em parquet 
```

```scala
df = spark.read.parquet("bronze/ingest_ibge_pib")
df.show(truncate=40) #Coloquei 40 só para mostrar o id 
```
